```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#open photographs in this notebook instead of another window
%matplotlib inline
```

### Create Classifier

```python
# define classifier
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
# open image with openCV
image = cv2.imread('Images/eye_face.jpg')
# openCV reads images with BGR scheme, but our pictures are RGB, so we convert
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# plot image
plt.imshow(fix_img)
```

### Detect Faces

```python
# 'detectMultiScale' finds features/locations of new image. Takes parameters a. gray scale variable, b. scaleFactor, c. minNeighbors
# a. we use the image properties, in this case RGB
# b. specifies how much the image size is reduced at each scale. this allows resizing of faces to increase chances of being detectable by the algo
# c. how any neighbors each rectangle should have to retain it. higher number = less detections but higher quality
faces = face_classifier.detectMultiScale(image, 1.3, 10)
```

### No Faces Detected

```python
# display a message if no faces are found
if faces is ():
    print('No faces found :(')
```

### Identify faces and draw rectangles

```python
def detect_face(fix_img):
    # create rectangles on faces using face_classifier with fix_img input
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    # define rectangle properties. x-pos, y-pos, width, height
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                      # bottom-left corner coordinates
                     (x,y),
                      # top-right corner coordinates
                     (x+w, y+h),
                      # color of rectangle
                     (255,0,0),
                      # line thickness
                     2
                     )
    return fix_img
```

```python
# store and print the result
result = detect_face(fix_img)
plt.imshow(result)
```
