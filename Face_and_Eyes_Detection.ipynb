# Face and Eyes Detection
### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#open photographs in this notebook instead of another window
%matplotlib inline
```

### Create Classifiers

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Open Image

```python
# open image with openCV
img = cv2.imread('Images/eye_face.jpg')
# openCV reads images with BGR scheme, but our pictures are RGB, so we convert
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

### Face Classifier

```python
faces = face_classifier.detectMultiScale(fix_img, 1.05, 1)
```

### If No Face Detected

```python
if faces is ():
    print('No faces found :(')
```

### Show Results

```python
def detect_faces_eyes(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    # define rectangles around the face
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                      (x,y),
                      (x+w, y+h),
                      (255,0,0),
                      2)
        # define rectangles around the eyes
        eyes_rects = eye_classifier.detectMultiScale(fix_img)
        for (ix,iy,iw,ih) in eyes_rects:
            cv2.rectangle(fix_img,
                        (ix,iy),
                        (ix+iw, iy+ih),
                        (0,0,255),
                        2)
    return fix_img
```

```python
# store and print the result
result = detect_faces_eyes(fix_img)
plt.imshow(result)
```
