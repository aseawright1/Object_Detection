# Car Plates Detection
### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Open Image

```python
img = cv2.imread('Images/car_plate2.jpg')
```

### Convert Image

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

### Display it

```python
display(img)
```

```python
plate_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

```python
def detect_plate(img):
    # create a copy of the image to define plate ontop of
    plate_img = img.copy()
    
    # define rectangles
    plate_rects = plate_classifier.detectMultiScale(plate_img, 1.1, 1)
    
    # define rectangle properties. x-pos, y-pos, width, height
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(plate_img,
                      # bottom-left corner coordinates
                     (x,y),
                      # top-right corner coordinates
                     (x+w, y+h),
                      # color of rectangle
                     (255,255,255),
                      # line thickness
                     7)
    return plate_img
```

### Check the Results

```python
result = detect_plate(img)
display(result)
```
