# Eyes Detection
### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#open photographs in this notebook instead of another window
%matplotlib inline
```

```python
### Open Image
```

```python
# open image with openCV
img = cv2.imread('Images/eye_face.jpg')
```

### Show Image

```python
# plot image
plt.imshow(img)
```

### Fix Image

```python
# openCV reads images with BGR scheme, but our pictures are RGB, so we convert
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

### Create Classifier

```python
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Detect Eyes in Image

```python
def detect_eyes(fix_img):
    # create rectangles on faces using face_classifier with fix_img input
    eyes_rects = eye_classifier.detectMultiScale(fix_img)
    
    # define rectangle properties. x-pos, y-pos, width, height
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(fix_img,
                      # bottom-left corner coordinates
                     (x,y),
                      # top-right corner coordinates
                     (x+w, y+h),
                      # color of rectangle
                     (255,255,255),
                      # line thickness
                     2
                     )
    return fix_img
```

### Show Results

```python
# store and print the result
result = detect_eyes(fix_img)
plt.imshow(result)
```
